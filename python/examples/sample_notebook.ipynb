```python
import qsa
from qsa import *
```

```python
e = qsa.rtxi.Experiment('sample_rtxi.json')
m = e.get_measurement(0)
a = qsa.analysis.Analysis(m)
```

```python
m.frequencies
```

```python
qsa.graphics.plot_comparison(a)
```

```python
qsa.graphics.figure()
qsa.graphics.subplot(2, 2, 1)
qsa.graphics.plot_qmatrix(a, new_figure=False)
qsa.graphics.subplot(2, 2, 2)
qsa.graphics.plot_qeigen(a, new_figure=False)
qsa.graphics.subplot(2, 2, 3)
qsa.graphics.plot_transfer(a, new_figure=False)
qsa.graphics.subplot(2, 2, 4)
qsa.graphics.plot_flat(a, legend=False, new_figure=False)
```
